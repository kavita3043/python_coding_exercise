```python
#import pandas library and check pandas version
import pandas as pd
pd.__version__
```

```python
# create a dataframe from a dictionary and display it.
#add index values
#add specified rows and columns from the dataframe (extract score and attempt columns where attempts is greater than 1)

d={'name':['kavi','mahi','ravi','savi','navi'],'score':[23,41,26,44,12],'attempts':[2,1,3,1,2]}

df=pd.DataFrame(d)
#print(df)


df=pd.DataFrame(d, index=['a','b','c','d','e'])
print(df)

df.loc[df['attempts']>1,['score','attempts']]
```

```python
#select the rows where the score is missing, i.e. is NaN.
d={'name':['kavi','mahi','ravi','savi','navi'],'score':['Nan',41,'Nan',44,12],'attempts':[2,1,3,1,2]}

df=pd.DataFrame(d)
df

df[df['score']=='Nan']
```

```python
#select the rows the score is between 15 and 20 (inclusive).
d={'name':['kavi','mahi','ravi','savi','navi'],'score':[18,41,16,44,20],'attempts':[2,1,3,1,2]}

df=pd.DataFrame(d)
df[df['score'].between(15,20)]

#select the rows where number of attempts in the examination is less than 2 and score greater than 15.
```

```python
#select the rows where number of attempts in the examination is less than 2 and score greater than 15.

df[(df['attempts']<2) & (df['score']>15)]
```
