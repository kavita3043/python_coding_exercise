```python
import pandas as pd
```

```python
# remove first 2 rows of a given DataFrame.
d={'name':['kavi','ravi','navi','savi','javi'],'age':[10,12,10,11,10],'score':[32,33,23,41,25],'qualify':['yes','yes','no','yes','no']}
df=pd.DataFrame(d)
df.iloc[2:]
```

```python
#remove last 2 rows of a given DataFrame.
df.iloc[:-2]
```

```python
#add a prefix or suffix to all columns of a given DataFrame.
df.add_suffix("_A")
df.add_prefix("A_")
```

```python
#reverse order (rows, columns) of a given DataFrame.

#rows
df.loc[::-1]

#columns
df.loc[:,::-1]
```

```python
#select columns by data type of a given DataFrame.
df.select_dtypes(include='number')

df.select_dtypes(include='object')
```

```python
#get lowest/largest 3 records within each numerical column of a given DataFrame.

#smallest
df.nsmallest(3,'score')
df.nsmallest(3,'age')

#largest
df.nlargest(3,'score')
df.nlargest(3,'age')
```

```python
#check for inequality of two given DataFrames.
d={'name':['kavi','tavi','javi','savi','javi'],'age':[11,12,10,11,10],'score':[32,39,23,41,25],'qualify':['yes','yes','no','yes','no']}
df1=pd.DataFrame(d)

df==df1
```

```python
#display the default index and set a column as an Index in a given dataframe.
df.index
temp=df.set_index('name')
temp
```

```python
#convert the string values to upper, lower cases in a given pandas series. Also find the length of the string values.
df['name'].str.upper()
df['name'].str.lower()
df['name'].str.len()
```
