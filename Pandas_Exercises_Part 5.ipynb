```python
import pandas as pd
```

```python
#join the two given dataframes along rows and assign all data.
d1={'name':['kavi','ravi','navi','savi','javi'],'age':[10,12,10,11,10],'score':[32,33,23,41,25],'qualify':['yes','yes','no','yes','no']}
df1=pd.DataFrame(d1)

d2={'name':['sonu','monu','tony','sam','leena'],'age':[10,22,15,11,10],'score':[35,13,13,42,25],'qualify':['yes','no','no','yes','no']}
df2=pd.DataFrame(d2)

df=pd.concat([df1,df2], ignore_index=True)
df
```

```python
#join the two given dataframes along columns and assign all data.
df_new=pd.concat([df1,df2], axis=1)
df_new
```

```python
# join the two dataframes using the common column of both dataframes.
d1={'name':['kavi','ravi','navi','savi','javi'],'age':[10,12,10,11,10],'score':[32,33,23,41,25],'qualify':['yes','yes','no','yes','no']}
df3=pd.DataFrame(d1)

d2={'name':['sonu','ravi','tony','savi','leena'],'age':[10,22,15,11,10],'score':[35,13,13,42,25],'qualify':['yes','no','no','yes','no']}
df4=pd.DataFrame(d2)

df_new=pd.merge(df3,df4, how='inner',on='name')
df_new
```

```python
#join (left join) the two dataframes using keys from left dataframe only.
df_new=pd.merge(df3,df4, how='left',on='name')
df_new
```

```python
#join two dataframes using keys from right dataframe only.
df_new=pd.merge(df3,df4, how='right',on='name')
df_new
```
